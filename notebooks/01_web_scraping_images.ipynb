```python
from duckduckgo_search import DDGS
import requests, os
from PIL import Image
from io import BytesIO

os.makedirs("cricket_images", exist_ok=True)

query = "Images of mens cricket international matches involving ball, bat and stump site:espncricinfo.com"
count = 500  # how many you want

with DDGS() as ddgs:
    results = ddgs.images(query, max_results=50)
    for i, r in enumerate(results, start=1):
        img_url = r["image"]
        try:
            img_data = requests.get(img_url, timeout=10).content
            img = Image.open(BytesIO(img_data)).convert("RGB")
            img = img.resize((800, 600))
            img.save(f"cricket_images/cricket_{i}.jpg")
            print(f"‚úÖ Saved image {i}")
        except Exception as e:
            print(f"‚ö†Ô∏è Skipped {i}: {e}")

print("üéØ All done! Images saved in cricket_images/")
```

```python
from duckduckgo_search import DDGS
import time

def search_and_download_images(term, max_images=150):
    print(f"Searching for '{term}' with a target of {max_images} images.")
    image_urls = []
    with DDGS() as ddgs:
        # The generator yields dictionaries with image information
        search_results_generator = ddgs.images(keywords=term)

        try:
            for _ in range(max_images): # Iterate up to max_images times
                result = next(search_results_generator)
                if result and 'image' in result:
                    image_urls.append(result['image'])
                else:
                    print("No more image results found for this query.")
                    break
                time.sleep(0.1) # Add a small delay between requests
        except StopIteration:
            print("Reached end of search results.")
        except Exception as e:
            print(f"An error occurred: {e}")

    print(f"Found {len(image_urls)} image URLs.")
    # You would then add code here to download these URLs
    return image_urls

# Example usage
urls = search_and_download_images("Images of mens cricket international matches involving ball, bat and stump site:espncricinfo.com", max_images=100)
# print(urls) # Uncomment to see the list of URLs
```
