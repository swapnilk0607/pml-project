```python
from duckduckgo_search import DDGS
import requests, os
from PIL import Image
from io import BytesIO

os.makedirs("cricket_ground_images", exist_ok=True)

query = "Images of cricket bat with no watermark"
count = 200  # how many you want

with DDGS() as ddgs:
    results = ddgs.images(query, max_results=count)
    for i, r in enumerate(results, start=1):
        img_url = r["image"]
        try:
            img_data = requests.get(img_url, timeout=10).content
            img = Image.open(BytesIO(img_data)).convert("RGB")
            img = img.resize((800, 600))
            img.save(f"cricket_ground_images/ground_{i}.jpg")
            print(f"âœ… Saved image {i}")
        except Exception as e:
            print(f"âš ï¸ Skipped {i}: {e}")

print("ğŸ¯ All done! Images saved in cricket_ground_images/")
```
