We can also show the sketch a second time taking into account the new values:

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import time
import os

# Set up the WebDriver (make sure you have the appropriate driver for your browser)
# For Chrome, download ChromeDriver from https://chromedriver.chromium.org/downloads
driver = webdriver.Chrome()

try:
    # Navigate to the website
    driver.get("https://www.espncricinfo.com/photo") # Replace with the actual URL
    time.sleep(10)
    # Locate the image element (replace with the correct locator for your image)
    # Example using CSS selector:
    image_element = driver.find_element(By.CSS_SELECTOR, "img.ds-rounded-xl")
    
    # Get the image URL from the 'src' attribute
    image_url = image_element.get_attribute("src")

    # Define the local file path to save the image
    save_directory = "downloaded_images"
    os.makedirs(save_directory, exist_ok=True)
    image_filename = "downloaded_image.jpg" # You can extract filename from URL or use a custom name
    save_path = os.path.join(save_directory, image_filename)

    # Download the image using requests
    if image_url:
        print(f"Downloading image from: {image_url}")
        response = requests.get(image_url, stream=True)
        response.raise_for_status() # Raise an exception for HTTP errors

        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=8192):
                file.write(chunk)
        print(f"Image saved to: {save_path}")
    else:
        print("Image URL not found.")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser
    driver.quit()
```
